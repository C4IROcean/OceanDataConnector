```python
import odp
import warnings
import lusedata_functions as ls
warnings.filterwarnings("ignore")
```

```python
db=odp.Database()
```

## Pull lusedata and create filter
### Filters can be for any column (i.e. location, date, lice count)

```python
filter1 = db.filter_data('dato', '>', '2018-01-01')
df = db.query('lusedata', filters=[filter1])
df.head()
```

```python
lice_type = ['voksne_hunnlus','fastsittende_lus','bevegelige_lus']
```

### The following function will create a dataframe for average lice counts as well as a plot

```python
df_avg, chart = ls.average_lice(df, [2019,2022], [1,52], lice_type[0], 'omrade', 'Hardanger', y_scale=None)
```

### The following function will create a dataframe comparing mean lice counts between locations, as well as a plot

```python
df_comp, chart = ls.average_lice_compare(df, 2020, [1,52], lice_type[0], 'omrade', ['Hardanger','Nordhordland','Senja'], y_scale=None)
```
